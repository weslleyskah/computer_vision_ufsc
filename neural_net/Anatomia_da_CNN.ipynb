# Baixando datast do Kaggle

Bixando dataset cats and dogs, o kaggle da o comando no site



```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("bhavikjikadara/dog-and-cat-classification-dataset")

print("Path to dataset files:", path)
```

# Analisando dataset

definindo nossos paths e analisando nosso dataset

```python
import os

data_dir = os.path.join(path, 'PetImages')
dog_path = os.path.join(data_dir, 'Dog')
cat_path = os.path.join(data_dir, 'Cat')


dog_imgs = os.listdir(dog_path)
cat_imgs = os.listdir(cat_path)

dog_imgs = [os.path.join(dog_path, img) for img in dog_imgs]
cat_imgs = [os.path.join(cat_path, img) for img in cat_imgs]

print(cat_imgs[:5])

len(os.listdir(dog_path))



```

# Separando Dataset

Fazendo a separaГ§ГЈo de train e val, ja que o kaggle nГЈo disponibiliza ele baixado para nГіs

```python
import tensorflow as tf
from tensorflow import keras

train_ds, val_ds = keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    image_size=(128,128),
    seed=123,
    subset="both",
    batch_size=32

)

classes_names = train_ds.class_names
print(classes_names)

train_ds = train_ds.apply(tf.data.experimental.ignore_errors())
val_ds = val_ds.apply(tf.data.experimental.ignore_errors())

train_ds_fast = train_ds.take(300)
val_ds_fast = val_ds.take(60)
```

# Criando o modelo

Declarando nossa rede com camadas convolutivas

```python

from keras import models, layers

model = models.Sequential([

    # Rede Convolucional
    layers.Input(shape=(128,128,3)), # camada de input ( linhas, colunas, dimensГµes)
    layers.Rescaling(1./255), # normalizaГ§ГЈo para rede aprender mais rГЎpido

    layers.Conv2D(16, 3, padding='same', activation='relu') , # primeira camada convolutiva
    layers.BatchNormalization(),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),

    layers.Conv2D(16, 3, padding='same', activation='relu') , # camada convolutiva
    layers.BatchNormalization(),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),

    layers.Conv2D(32, 3, padding='same', activation='relu') , # camada convolutiva
    layers.BatchNormalization(),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),

    layers.Conv2D(64, 3, padding='same', activation='relu') , # camada convolutiva
    layers.BatchNormalization(),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),

    layers.Conv2D(512, 3, padding='same', activation='relu') , # camada convolutiva
    layers.BatchNormalization(),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),

    layers.Conv2D(1024, 3, padding='same', activation='relu') , # camada convolutiva
    layers.BatchNormalization(),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),

    layers.Conv2D(2048, 3, padding='same', activation='relu') , # camada convolutiva
    layers.BatchNormalization(),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),

    # Rede Neural
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dense(2,activation='softmax')

])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(
    train_ds_fast,
    validation_data=val_ds_fast,
    epochs=48,
    batch_size=64,
)
```

```python
test_loss, test_accuracy = model.evaluate(val_ds)
```
