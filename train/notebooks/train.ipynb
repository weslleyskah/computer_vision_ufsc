## Training

```python
from ultralytics import YOLO
from pathlib import Path

model = YOLO("yolo26n-seg.pt")

dataset_yaml_path = "/content/input/data.yaml"

# Paths
ROOT = Path.cwd()
OUTPUT_DIR = ROOT / "output"

# Config names
model_variant = "yolo26n-seg" # Model
img_size = "640"
epochs = "500"
batch_size = "64"

run_name = f"{model_variant}_{img_size}_e{epochs}_b{batch_size}"

results = model.train(
    data=dataset_yaml_path,
    epochs=500,
    imgsz=640,
    batch=64,
    project=str(OUTPUT_DIR),  # Output: /content/output/
    name=run_name             # Folder: /content/output/yolo26n-seg_640_e500_b64/
)
```
