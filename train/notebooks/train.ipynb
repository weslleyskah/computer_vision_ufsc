## Training

```python
from google.colab import drive
from ultralytics import YOLO
from pathlib import Path

drive.mount('/content/drive')

ROOT = Path.cwd()
OUTPUT_DIR = ROOT / "output"
dataset_yaml_path = "/content/input/data.yaml"

best_model = "/content/drive/MyDrive/Models/fireseg_v2.pt"
model = YOLO(best_model)
```

```python
%%writefile /content/input/data.yaml
path: /content/input
train: train/images
val: train/images

nc: 2
names: ['fire', 'smoke']
```

```python
!find /content/input/ -name "*.txt" -exec sed -i '/^1 /d' {} +
```

```python
# Config
model_variant = "yolo26n-seg" # initial model
img_size = "640"
epochs = "500"
batch_size = "64"

run_name = f"{model_variant}_{img_size}_e{epochs}_b{batch_size}_"

# Run
results = model.train(
    data=dataset_yaml_path,
    epochs=500,
    imgsz=640,
    batch=64,
    project=str(OUTPUT_DIR),  # Output: /content/output/
    name=run_name,             # Folder: /content/output/yolo26n-seg_640_e500_b64/
    val=False
    )
```
